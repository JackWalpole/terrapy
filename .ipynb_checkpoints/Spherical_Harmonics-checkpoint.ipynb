```python
# for SHTOOLS
import pyshtools as shtools
lmax = 10
cilm, chi2 = shtools.expand.SHExpandLSQ(A.vp[5],A.lat,A.lon,lmax)
grid = shtools.expand.MakeGrid2D(cilm)                                        
                                        
                                        
```

```python
# map = Basemap(projection='kav7',lon_0=180,resolution='c')
map = Basemap(projection='mill',lon_0=180,resolution='c')
nx = grid.shape[1]
ny = grid.shape[0]
lons, lats = map.makegrid(nx,ny)
x, y =map(lons,-lats)
map.contourf(x,y,grid,cmap='RdYlBu')
map.drawcoastlines()
plt.colorbar()
plt.show()

```
